```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_player_url(): 
    player_lkp = {}
    for year in range(2009, 2020 + 1):       
        URL = 'https://www.nbadraft.net/ranking/bigboard/?year-ranking=' + str(year)
        text = requests.request(url = URL, method = 'GET').text
        soup = BeautifulSoup(text, 'html.parser')
        table = soup.find({'table' : 'DataTables_Table_0'})
        rows = table.findAll('td')
        for counter, row in enumerate(rows): 
            row = str(row)
            if 'player' in row: 
                first_name = row.split('<span>')[1].split('<')[0]
                last_name = row.split('<span>')[2].split('<')[0]
                player = first_name + ' ' + last_name
                url = row.split('href="')[1].split('"')[0]
                player_lkp[player] = url
    
    return player_lkp
```

```python
players = get_player_url()
```
